# Secara Visual

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
sales = pd.read_csv('D:\CSV\sales-feb-2015.csv')
```

```python
sales.shape
```

```python
sales.head()
```

```python
sales.plot(x='Date', y='Units')
```

```python
sales.plot(x='Date', y='Units', kind='scatter')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Units')
```

```python
sales.plot(y='Units', kind='box')
plt.ylabel('sepal width (cm)')
```

```python
sales.plot(y='Units', kind='hist')
plt.xlabel('Dates')
```

```python
sales.plot(y='Units', kind='hist', bins=30, range=(0,20), density=True)
plt.xlabel('Dates')
plt.show()
```

```python
sales.plot(y='Units', kind='hist', bins=30, range=(0,20), cumulative=True, density=True)
plt.xlabel('Dates')
plt.title('Cumulative distribution function (CDF)')
```
